# Multiple Images
**In this example, an image of an eye is received along with a mask designed for it. After image normalization, two green and blue channels are removed from the eye image and the red channel is masked.**

## Imports

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

## Normalizer and Denormalizer functions

```python
normalize = lambda inArr: (inArr - inArr.min()) / (inArr.max() - inArr.min())
deNormalize = lambda img: np.array(img * 255, dtype=np.uint8)
```

## Load Images and conver them to RGB

```python
image_eye=cv2.imread('imgs/eye.jpg')
image_eye=cv2.cvtColor(image_eye,cv2.COLOR_BGR2RGB)

image_eye_mask=cv2.imread('imgs/eye-mask.png')
image_eye_mask=cv2.cvtColor(image_eye_mask,cv2.COLOR_BGR2RGB)
```

## Normalize Images

```python
image_eye=normalize(image_eye)
image_eye_mask=normalize(image_eye_mask)
```

### Initialize matplotlib

```python
figure=plt.figure()
```

### Show eye image

```python
figure.add_subplot(2,2,1)
plt.imshow(image_eye)
plt.axis('off')
plt.title('Original Image')
```

### Show mask 

```python
figure.add_subplot(2,2,2)
plt.imshow(image_eye_mask)
plt.axis('off')
plt.title('Mask Image')
```

## Make masked image
**Remove the two green and blue layers by zeroing them and multiplying the red layer by the mask**

```python
x=np.zeros((image_eye.shape[0],image_eye.shape[1]),dtype=float)
output=np.stack((image_eye_mask[...,0]*image_eye[...,0],x,x),axis=2)
```

### Show masked image

```python
figure.add_subplot(2,2,3)
plt.imshow(output)
plt.axis('off')
plt.title('Masked Image')
```

### Show initialized plt

```python
plt.show()
```
