```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = 255*np.ones((400, 400, 3), dtype=np.uint8)

# main purple ellipse
cv2.ellipse(img, center=(200, 200), axes=(150, 190), angle=0 , startAngle=0, endAngle=360, color=(141, 0, 67), thickness=-1)

# ellipse to color part of main ellipse
cv2.ellipse(img, center=(200, 200), axes=(150, 191), angle=0 , startAngle=-90, endAngle=-18, color=(2, 171, 255), thickness=-1)

# top circle
cv2.circle(img, center=(200, 150), radius=100, color=(255, 255, 255), thickness=-1)

# top rectangle
cv2.rectangle(img, pt1=(190, 9), pt2=(210, 50), color=(255, 255, 255), thickness=-1)

# right rectangle
cv2.rectangle(img, pt1=(250, 140), pt2=(350, 160), color=(255, 255, 255), thickness=-1)

# bottom cicrle
cv2.circle(img, center=(200, 390), radius=70, color=(255, 255, 255), thickness=-1)

plt.imshow(img[..., ::-1])
plt.axis('off')
plt.show()
```
